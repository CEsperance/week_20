# Stop Words

```python
#remove punctuation
from string import punctuation
print(punctuation)
no_punct = [w for w in pl_words if w not in punctuation]

#remove stop words
from nltk.corpus import stopwords

sw1 = stopwords.words('english')
print(sw1)
no_sw1 = [w for w in no_punct if w.lower() not in sw1]
```

```python
#stemming
from nltk.stem.snowball import SnowballStemmer

#instatiate snowball stemmer
stemmer = SnowballStemmer(language='english')
```

```python
for word in no_sw1:
    print(word, stemmer.stem(word))
```

## Lemmentizer

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

for word in no_sw1:
    print(word, lemmatizer.lemmatize(word))
```

**For this example, I prefer lemmatization because it is more accurate in bringing the words to their base form. For example, in the stemming process, it changed disobedience to disobedi and silence to silenc. Lemmatization recongnizes the base word in context and doesn't remove letters that are needed.**

```python
#continue to remove meaningless words with more preprocessing
unimportant = ['thy', 'thus', 'thou', 'thee', 'yet', 'and']

final = [l for l in no_sw1 if l not in unimportant]
```

```python
#print the occurrence of each word

from nltk.probability import FreqDist
vocab = nltk.FreqDist(final)
```

```python
top_20 = vocab.most_common(20)
top_20
```

```python
#https://dariuslfuller.medium.com/creating-visuals-with-nltks-freqdist-ac4e667e49f3

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#convert the frequency to a pandas series via dictionary for easier plotting
top_20 = pd.Series(dict(top_20))

#set figure and ax into variables
fig,ax = plt.subplots(figsize = (10,10))

all_plot = sns.barplot(x = top_20.index, y = top_20.values, ax = ax)
plt.xticks(rotation=30)
```

# 4. Perform Vader Sentiment Analysis on the book. Find the 5 most negative, 5 most positive, and 5 most neutral sentences in Paradise Lost. http://www.nltk.org/howto/sentiment.html . This may take a while to run, so you can always start with a small subset of the data (100 sentences) and then once your code works as expected, expand it to the whole book and let it run. 

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
analyzer = SentimentIntensityAnalyzer()
```

```python
notebook = nltk.corpus.gutenberg.open('milton-paradise.txt')
```

```python
#sentiment = analyzer.polarity_scores(notebook)
```

```python
# a list for the sentiments
sentiments = []


for sentences in notebook:
    sentiment = analyzer.polarity_scores(sentences)
    compound = sentiment["compound"]
    pos = sentiment['pos']
    neu = sentiment['neu']
    neg = sentiment['neg']

    sentiments.append({
                "text":sentences,
                "compound":compound,
                "positive":pos,
                "negative":neg,
                "neutral":neu
    })
    
sentiment_df = pd.DataFrame(sentiments)
```

```python
positive = sentiment_df[['positive', 'text']]
five_positive = positive.sort_values(ascending = False, by = 'positive')
five_positive.head()
```

```python
negative = sentiment_df[['negative', 'text']]
five_negative = negative.sort_values(ascending = False, by = 'negative')
five_negative.head()
```

```python
neutral = sentiment_df[['neutral', 'text']]
five_neutral = neutral.sort_values(ascending = False, by = 'neutral')
five_neutral.head()
```
